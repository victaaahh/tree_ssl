```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import RegularPolygon, Circle
from shapely.geometry import Point
import geopandas as gpd
```

```python
xy_radius = 15
x_min = 0
x_max = 188
y_min = 0
y_max = 183

hex_grid_dist = np.sqrt(3*xy_radius*xy_radius)

x_coords = np.arange(x_min+xy_radius, x_max-xy_radius, 1.5*xy_radius)
y_coords = np.arange(y_min+hex_grid_dist/2, y_max-hex_grid_dist/2, hex_grid_dist)
```

```python
#point_list = []
x_list = []
y_list = []
for y in y_coords:
    for x_idx, x in enumerate(x_coords):
        #point_list.append(Point(x+xy_radius*(y_idx%2),y))
        if x_idx%2==0:
            x_list.append(x)
            y_list.append(y)
        elif y_max-y >= hex_grid_dist:
            x_list.append(x)
            y_list.append(y+hex_grid_dist/2)

#labels = gpd.GeoDataFrame(geometry=point_list)
```

```python
plt.figure(figsize=(15,15))
plt.scatter(x_list, y_list)
plt.axvline(x_min)
plt.axvline(x_max)
plt.axhline(y_min)
plt.axhline(y_max)
plt.xlim(x_min-15, x_max+15)
plt.ylim(y_min-15, y_max+15)

for x, y in zip(x_list, y_list):
    hexagon = RegularPolygon((x, y), numVertices=6, radius=xy_radius, orientation=np.pi/6, fill=False)
    circle = Circle((x,y), radius=xy_radius, fill=False)
    plt.gca().add_patch(hexagon)
    plt.gca().add_patch(circle)
```
